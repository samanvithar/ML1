```python
import pandas as pd
from sklearn.svm import SVC,SVR
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('UniversalBank.csv')
```

```python
df.head()
```

```python
df['CreditCard'].value_counts()
```

```python
X=df.drop(columns=['Personal Loan','ID','ZIP Code'],axis=1)
y=df['Personal Loan']
```

```python
Scalar=StandardScaler()
scal=Scalar.fit_transform(X)
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(scal,y,test_size=0.2,random_state=42)
```

```python
model=SVC()
model.fit(Xtrain,ytrain)
```

```python
pred=model.predict(Xtest)
```

```python
confusion_matrix(ytest,pred)
```

```python
f1_score(ytest,pred,pos_label=1)
```

```python
param={
    'kernel':['rbf'],
    'C':[0.01,0.1,1,10,100],
    'gamma':[0.001,0.01,0.1,1,'scale']
}
```

```python
grid=GridSearchCV(model,param,cv=5,n_jobs=-1)
```

```python
grid.fit(Xtrain,ytrain)
```
