```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('loan_approval_dataset.csv')
df.head()
```

```python
df.drop('Loan_Term',axis=1,inplace=True)
```

```python
X=df.drop('Loan_Approved',axis=1)
y=df['Loan_Approved']

Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,train_size=0.8,random_state=42)
cat_col=X.select_dtypes(include='object').columns
num_col=X.select_dtypes(include=['int','Float64']).columns
```

```python
preprocessor=ColumnTransformer(
          transformers=[('OneHotEncoder',OneHotEncoder(sparse_output=False),cat_col)],remainder='passthrough'
)
```

```python
pipeline=Pipeline(
           steps=[('preprocessing',preprocessor),
                  ('decision',DecisionTreeClassifier(random_state=42,criterion='entropy'))
               ]
)
```

```python
pipeline.fit(Xtrain,ytrain)
```

```python
pipeline.score(Xtest,ytest)
```
