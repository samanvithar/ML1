```python
from sklearn.model_selection import train_test_split,GridSearchCV
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/santhoshkumaroff/MachineLearningDatasets/refs/heads/main/dataset_traffic_accident_prediction1.csv')
```

```python

df_clean = df[df['Accident'].notnull()]


X=df_clean.drop('Accident',axis=1)
y=df_clean['Accident']
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,random_state=42,train_size=0.8)





```

```python
cat_cols=X.select_dtypes(include='object').columns
num_cols=X.select_dtypes(exclude='object').columns
```

```python
num_pipeline = Pipeline(
    steps=[
        ('num_impute',SimpleImputer(strategy="mean")),
        ('scalar', StandardScaler())
    ]
)

cat_pipeline = Pipeline(
    steps=[
        ('cat_impute',SimpleImputer(strategy='most_frequent')),
        ('encode',OneHotEncoder(sparse_output=False))
    ]
)


preprocessor = ColumnTransformer(
    transformers=[
        ('num',num_pipeline,num_cols),
        ('cat',cat_pipeline,cat_cols)
    ]
)



pipe = Pipeline(
    steps=[
        ('preprocessing', preprocessor),
        ('model', RandomForestClassifier(random_state=42,n_estimators=100))
    ]
)

pipe.fit(Xtrain, ytrain)


```

```python
pipe.score(Xtest,ytest)
```

```python
pipe.score(Xtrain,ytrain)
```

# GridsearchCV

```python
params={
    'model__criterion':['entropy','gini'],
    'model__min_samples_split':[2,3,4,5,10],
    'model__min_samples_leaf':[2,3,5,7,10],
    'model__max_depth':[5,10,50,100,200]
}
```

```python
gridSearch=GridSearchCV(pipe,params,cv=5,n_jobs=-1)
```

```python
gridSearch.fit(Xtrain,ytrain)
```

```python
gridSearch.score(Xtest,ytest)
```

```python
gridSearch.score(Xtrain,ytrain)
```
